```python
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing import event_accumulator
import numpy as np
import os
```

```python
dir_path = '/Users/vincent/vincent/learn/Research/drl_iot/RL_IoT_distillation/vanila/log/vanilla'
logs = {}
for path, dir_list, file_list in os.walk(dir_path):
    # print(path, dir_list, file_list)
    # print(dir_list)
    for file_name in file_list:
        file_path = os.path.join(path, file_name)
        if 'events' in file_path:
            env_name = file_path.split('/')[-4]
            logs[env_name] = file_path
# ea = event_accumulator.EventAccumulator(tfevent_file)
# ea.Reload()
# print(ea.scalars.Keys())
```

```python
def get_mean_last10(tfevent_file):
    ea = event_accumulator.EventAccumulator(tfevent_file)
    ea.Reload()
    test_rew = ea.scalars.Items('test/rew')
    test_rew_value = [i.value for i in test_rew]
    return np.mean(test_rew_value[-10:])
```

```python
for name, path in logs.items():
    print('env_name:{},reward:{}'.format(name, get_mean_last10(path)))
```

```python
def normal(nums):
    mu = np.average(nums)
    sigma = np.std(nums)
    return (nums - mu) / sigma
```
