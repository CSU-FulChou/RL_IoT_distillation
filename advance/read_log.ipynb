```python
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing import event_accumulator
import numpy as np
```

```python
tfevent_file = '/home/zhoufu/log/Ant-v3/sac/seed_0_1020_101151-Ant_v3_always_pd_64'
ea = event_accumulator.EventAccumulator(tfevent_file)
ea.Reload()
print(ea.scalars.Keys())
```

```python
def normal(nums):
    mu = np.average(nums)
    sigma = np.std(nums)
    return (nums - mu) / sigma
```

```python
test_rew = ea.scalars.Items('test/rew')
test_rew_std = ea.scalars.Items('test/rew_std')
# loss_kl_loss = ea.scalars.Items('loss/kl_loss')
# loss_actor = ea.scalars.Items('loss/actor')
# loss_critic1 = ea.scalars.Items('loss/critic1')
# loss_critic2 = ea.scalars.Items('loss/critic2')

plt.figure(figsize=(8, 8), dpi=180)
plt.figure(1)
# plt1 = plt.subplot(221)
rew = plt.plot([i.step for i in test_rew], normal([i.value for i in test_rew]), label='rew')
# plt2 = plt.subplot(222)
# kl_loss = plt.plot([i.step for i in loss_kl_loss], normal([i.value for i in loss_kl_loss]), label='kl_loss')
# actor = plt.plot([i.step for i in loss_actor], normal([i.value for i in loss_actor]), label='actor')
# critic1 = plt.plot([i.step for i in loss_critic1], normal([i.value for i in loss_critic1]), label='critic1')
rew_std = plt.plot([i.step for i in test_rew_std], normal([i.value for i in test_rew_std]), label='rew_std')

plt.legend()
plt.show()
# 'loss/actor', 'loss/critic1', 'loss/critic2',
```
